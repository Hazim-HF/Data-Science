```python
# read sales and product files
sales = pd.read_csv('./Data/PartB_Q4/sales_data.csv')
product = pd.read_csv('./Data/PartB_Q4/products.csv')
```

```python
sales.head()
```

```python
product.head()
```

```python
sales_product = pd.merge(sales, product, on = 'Product_ID', how = 'inner')
sales_product.head()
```

```python
# calculate total sales
sales_product['total_sales'] = sales_product['Quantity']*sales_product['Price']

# calculate total sales for each product
total_sales = sales_product.groupby('Product_Name')['total_sales'].sum()
total_sales
```

```python
# calculate the average sold quantity for each product
average_quantity = sales_product.groupby('Product_Name')['Quantity'].mean()
average_quantity
```

```python
# plot the total sales for each product
import matplotlib.pyplot as plt
total_sales.plot(kind='bar')
plt.show()
```

### c - 

```python
customer = pd.read_csv('./Data/PartB_Q4/customers.csv')
customer.head()
```

```python
merged2 = pd.merge(sales_product, customer, on = 'Customer_ID', how = 'inner')
```

```python
# calculate number of transaction based on country
num_transaction = merged2.groupby('Country')['Transaction_ID'].nunique()
num_transaction
```

```python
# calculate total sales based on country
total_sales_country = merged2.groupby('Country')['total_sales'].sum()
total_sales_country
```

```python
type(total_sales_country)
```

```python
df = total_sales_country.to_frame().sort_values(by = 'total_sales', ascending=False)
df
```

```python
df['Country'] = df.index
df
```

```python
df.reset_index(drop = True, inplace = True)
df
```

```python
# Interpolate
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {'A': [1, np.nan, 3, np.nan, 5], 'B': [2, 4, np.nan, 8, 10]}
df = pd.DataFrame(data)

# Apply interpolation
df_interpolated = df.interpolate()

print(df_interpolated)
```

```python
#These methods are useful for smooth interpolations, especially for time series data.
df.interpolate(method='polynomial', order=2)  # Quadratic interpolation
df.interpolate(method='spline', order=3)      # Cubic spline interpolation
```
