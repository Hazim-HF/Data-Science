```python
class GeneralTracker:
    """
    Class to track basic financial transactions (deposit and withdrawal) and display summary.
    """
    def __init__(self, acc_holder):
        self.acc_holder = acc_holder
        self.bal = 0
        self.record = []  

    def deposit(self, amt, inc_type):
        
        if amt <= 0:
            raise ValueError('You must deposit at least RM 0.01.')
        self.bal += amt
        self.record.append((amt, inc_type, 'Income'))
        print(f'You have successfully deposited RM {amt:.2f} ({inc_type}) into your account.')

    def withdraw(self, amt, expense_type):
        
        if amt <= 0:
            raise ValueError('Withdrawal amount must be greater than RM0.00.')
        if amt > self.bal:
            raise ValueError('Withdrawal rejected. Reason: Insufficient balance.')
        self.bal -= amt
        self.record.append((amt, expense_type, 'Expense'))
        print(f'You have successfully withdrew RM {amt:.2f} ({expense_type}) from your account.')

    def acc_summary(self):
        print(f'\n<<<<< Financial Summary for {self.acc_holder} >>>>>')
        print(f'Account Balance: RM {self.bal:.2f}')

        print('\nTransaction Records:')
        if self.record:
            for idx, (amt, type_desc, category) in enumerate(self.record, 1):
                print(f'{idx}. RM {amt:.2f} ({type_desc}, {category})')
        else:
            print('No transaction recorded.')


class PersonalTracker(GeneralTracker):
    """
    Class to track personal financial transactions and savings goal.
    """
    def __init__(self, acc_holder):
        super().__init__(acc_holder)
        self.savings_goal = 0

    def set_savings_goal(self, goal):
        self.savings_goal = goal
        print(f'Savings goal set to RM {self.savings_goal:.2f}.')

    def check_savings_goal(self):
        if self.bal >= self.savings_goal:
            print(f'Congratulations! You have met your savings goal of RM {self.savings_goal:.2f}.')
        else:
            print(f'You are now only RM {self.savings_goal - self.bal:.2f} away from your savings goal.')


class BusinessTracker(GeneralTracker):
    """
    Class to track business financial transactions and expense limit.
    """
    def __init__(self, acc_holder):
        super().__init__(acc_holder)
        self.expense_limit = 0

    def set_expense_limit(self, limit):
        self.expense_limit = limit
        print(f'You have set your expenses limit to RM {self.expense_limit:.2f}.')

    def check_expense_limit(self):
        total_expenses = sum(amount for amount, _, category in self.record if category == 'Expense')
        if total_expenses > self.expense_limit:
            print(f'Warning! You have exceeded your expenses limit by RM {total_expenses - self.expense_limit:.2f}.')
        else:
            print(f'You are within your expense limit. Total Expenses: RM {total_expenses:.2f}.')

def simulation():
    print('\n<<<<< Financial Tracker Simulation with Inheritance >>>>>')

    pt = PersonalTracker(acc_holder='Samad Personal')
    pt.deposit(1700, 'Gaji')
    pt.deposit(200, 'Gig work')
    pt.withdraw(700, 'Sewa rumah')
    pt.withdraw(500, 'Sewa kereta')
    pt.set_savings_goal(1000)
    pt.check_savings_goal()
    pt.acc_summary()

    bt = BusinessTracker(acc_holder='Samad & Co.')
    bt.deposit(20000, 'Jualan')
    bt.withdraw(3000, 'Beli stok jualan')
    bt.withdraw(5000, 'Gaji pekerja')
    bt.set_expense_limit(9000)
    bt.check_expense_limit()
    bt.acc_summary()

simulation()
```
